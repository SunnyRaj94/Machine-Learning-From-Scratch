# Plotly Scatterplot - scatter plot

problem statement:
 Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
#importing plotly express for scatterplot
import plotly.express as ply
#importing numpy
import numpy as np , pandas as pd
#generating random x axis and y axis values and creating a dataframe out of it
values = pd.DataFrame({'Column1': np.random.randn(500), 'Column2': np.random.randn(500)})
#plotting graph with plotly
ply.scatter(values, x="Column1", y="Column2",color="Column2")
```
