# Plotly Scatterplot

problem statement:
 Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
#importing plotly express for scatterplot
import plotly.express as ply
#importing numpy
import numpy as np
#generating random x axis and y axis values
x_values = np.random.randn(1000)
y_values =np.random.randn(1000)
#plotting graph with plotly
ply.scatter(x=x_values, y=y_values)
```
