# Plotly Scatterplot -line plot

problem statement:
. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python
#importing plotly express for scatterplot
import plotly.graph_objects as ply
#importing numpy
import numpy as np
#generating random x axis and y axis values
x_values = np.random.randn(100)
y_values =np.random.randn(100)
#plotting graph with plotly
ply.Figure(ply.Scatter(x=x_values, y=y_values,mode='markers+lines'))
```
