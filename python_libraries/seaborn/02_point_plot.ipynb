#   Seaborn pointplot

problem statement:
Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
#importing matplotlib 
import matplotlib.pyplot as plt
#importing seaborn
import seaborn as sns
#import pandas
import pandas as pd
#reading csv file as data frame
titanic_df = pd.read_csv(open('titanic.csv'))
```

```python
#plotting graph from seaborn module by passing data frame
sns.pointplot(y='survived',x='sex',data=titanic_df)
```
