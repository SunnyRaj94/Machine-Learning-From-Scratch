# random distribution of different sizes balls

Problem Statement:
Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes. 

```python
#importing matplotlib
import matplotlib.pyplot as plt
#importing numpy
import numpy as np,random
#generating random x axis values 
x_axis_values = np.random.randn(25)
#generating random y axis values
y_axis_values = np.random.randn(25)
#taking some colors
color=['r','b','g','y']
#generating random colors 
colors = []
for i in range(25):
    i=color[random.randint(0,2)]
    colors.append(i)
#generating some random areas
area =[]
for i in range(25):
    area.append(random.randint(5,100))
#plotting scatterplot
plt.scatter(x_axis_values,y_axis_values,s=area, c=colors, alpha=0.85)
#providing lable to x and y axis
plt.xlabel("x_axis_values")
plt.ylabel("y_axis_values")
#showing graph
plt.show()
```
