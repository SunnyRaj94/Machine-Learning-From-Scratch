# creat scatterplot with random value of x and y in circle 

Problem statement:
Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.

```python
#importing matplotlib
import matplotlib.pyplot as plt
#importing numpy
import numpy as np
#generating random x axis values 
x_axis_values = np.random.randn(50)
#generating random y axis values
y_axis_values = np.random.randn(50)
#plotting scatterplot
plt.scatter(x_axis_values,y_axis_values, s=70, facecolors='none', edgecolors='b')
#providing lable to x and y axis
plt.xlabel("x_axis_values")
plt.ylabel("y_axis_values")
#showing graph
plt.show()
```
