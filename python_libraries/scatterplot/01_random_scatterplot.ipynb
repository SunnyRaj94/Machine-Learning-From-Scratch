# creat scatterplot with random value of x and y

problem statement:
Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.

```python
#importing matplotlib
import matplotlib.pyplot as plt
#importing random from pylab
from pylab import randn
#generating random x axis values 
x_axis_values = randn(200)
#generating random y axis values
y_axis_values = randn(200)
#plotting scatterplot
plt.scatter(x_axis_values,y_axis_values, color='r')
#providing lable to x and y axis
plt.xlabel("x_axis_values")
plt.ylabel("y_axis_values")
#showing graph
plt.show()
```
