# scatterplot of heights and weights

problem statement:
Write a Python program to draw a scatter plot for three different groups camparing weights and heights

```python
#importing matplotlib
import matplotlib.pyplot as plt
#initializing some random data
weight_one=[67,57.2,59.6,59.64,55.8,61.2,60.45,61,56.23,56]
height_one=[101.7,197.6,98.3,125.1,113.7,157.7,136,148.9,125.3,114.9] 
weight_two=[61.9,64,62.1,64.2,62.3,65.4,62.4,61.4,62.5,63.6]
height_two=[152.8,155.3,135.1,125.2,151.3,135,182.2,195.9,165.1,125.1] 
weight_three=[68.2,67.2,68.4,68.7,71,71.3,70.8,70,71.1,71.7]
height_three=[165.8,170.9,192.8,135.4,161.4,136.1,167.1,235.1,181.1,177.3]
#plotting first group of values
plt.scatter(weight_one, height_one, marker='*', color='red')
#plotting second group of values
plt.scatter(weight_two, height_two, color='green')
#plotting third group of values
plt.scatter(weight_three, height_three, s=50, color='blue')
#generating x label
plt.xlabel('weight', fontsize=16)
#generating y label
plt.ylabel('height', fontsize=16)
#giving figure a title
plt.title('Group wise Weight vs Height scatter plot',fontsize=20)
#showing output
plt.show()
```
