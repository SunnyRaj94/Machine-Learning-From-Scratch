# create a pie chart with a proper title

Problem Statement:
Write a Python programming to create a pie chart with a title of the popularity of programming Languages

```python
#importing matplotlib
import matplotlib.pyplot as plt
#initializing data as given in problem
programming_languages=['Java','Python','PHP','JavaScript','C#','C++']
popularity =[22.2,17.6,8.8,8,7.7,6.7]
#making a pie chart
plt.pie(popularity,labels = programming_languages, colors = 'rbgyk')
#giving it a title
plt.title("language popularity chart")
#displaying pie chart
plt.show()
```
