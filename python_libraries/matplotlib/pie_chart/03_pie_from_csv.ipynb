# create a pie chart , take data from csv file

Problem Statement:
Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file

```python
#importing matplotlib and panda
import matplotlib.pyplot as plt
import pandas as pd
#taking data from csv file which will generatew a data frame
csv_data = pd.read_csv(open('medal.csv'))
#plotting a pie by taking gold_medals as percentage and country name as its lable
plt.pie(csv_data['gold_medal'],labels = csv_data['country'], colors = 'rbgyk')
#showing the output
plt.show()
```
