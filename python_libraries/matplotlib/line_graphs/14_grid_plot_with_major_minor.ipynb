# Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid) and plot the graph

"""
Problem Statement : 
Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. 
between October 3, 2016 to October 7, 2016. 
Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).
Turn on the grid but turn off ticks
"""

```python
#importing matplotlib
import matplotlib.pyplot as plt
#importing pandas module
import pandas as pd
#setting data
data = [
        ["03-10-16",772.559998],
        ["04-10-16",776.429993],
        ["05-10-16",776.469971],
        ["06-10-16",776.859985],
        ["07-10-16",775.080017]
      ]
#extracting data into variable
x_axis_data = [row[0] for row in data]
y_axis_data = [row[1] for row in data]
#plotting graph date vs closing values with label
plt.plot(x_axis_data,y_axis_data ,label= 'Date vs Alphabet Inc Closing values')
# customising grid lines 
# major grid lines with dark grey lines
plt.grid(which='major', color='gray', ls='-', lw = .8)
# minor grid lines with faint grey lines
plt.grid(which='minor', color='gray', ls='--', lw = .5)
plt.minorticks_on() 
#adding title to the graph
plt.title("Closing values of Alphabet Inc stocks")     
#setting x_axis-label as Date
plt.xlabel('Date')
#setting y axis label as closing values
plt.ylabel('Closing values') 
```
