# program to create multiple types of charts

problem statement:
Write a Python program to create multiple types of charts

```python
#importing matplot lib and numpy
import matplotlib.pyplot as plt
import numpy as np
#generating a variable array
x = np.arange(0,0.5,0.05)
#plotting graph
plt.plot(x, x**2)
#plotting scatter plot 
plt.scatter(x, x**3)
#displaying graphs
plt.show()
#generating lables / values of list for pie chart
my_values = ['one','two','three','four','five','six','seven','eight','nine','ten']
#showing the graph
plt.show()
#plotting pie chart with above data
plt.pie(x, labels = my_values, colors = 'rbgyk')
#showing the output
plt.show()
```
