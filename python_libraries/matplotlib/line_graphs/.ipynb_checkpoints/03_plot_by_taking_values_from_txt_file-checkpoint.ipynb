### draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title

problem statement:
Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title

```python
#importing matplotlib
import matplotlib.pyplot as plt
#opening txt file
file = open('test.txt','r')
#reading data from txt file
data = file.readlines()
#putting values in x axis values 
x_axis_data = [int(line.split()[0]) for line in data]
# putting value in y axis data
y_axis_data = [int(line.split()[1]) for line in data]
```

```python
# giving both axix a label
plt.xlabel('this is data on my x axis')
# Set the y axis label of the current axis.
plt.ylabel('this is data on my y axis')
#giving it a proper label
plt.title('plot my awesome graph.')
#plotting graph from given values
plt.plot(x_axis_data,y_axis_data)
```
