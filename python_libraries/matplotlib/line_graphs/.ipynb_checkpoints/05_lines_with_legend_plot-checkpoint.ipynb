# plot two or more lines on same plot with suitable legends of each line

problem statement:
Write a Python program to plot two or more lines on same plot with suitable legends of each line

```python
#importing matplotlib
import matplotlib.pyplot as plt
# line 1 points
x_axis_line_1 = [10,20,30,70,80,90]
y_axis_line_1 = [20,40,10,23,22,12]
# plotting the line 1 points 
plt.plot(x_axis_line_1, y_axis_line_1, label = "my first line")
# line 2 points
x_axis_line_2 = [10,20,30]
y_axis_line_2 = [40,10,30]
# plotting the line 2 points 
plt.plot(x_axis_line_2, y_axis_line_2, label = "my second line")
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title of the current axes.
plt.title('Two or more lines on same plot with suitable legends ')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
