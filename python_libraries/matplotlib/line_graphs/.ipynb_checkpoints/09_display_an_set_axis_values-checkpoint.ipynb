# display the current axis limits values and set new axis values

problem statement:
Write a Python program to display the current axis limits values and set new axis values

```python
#importing matplotlib
import matplotlib.pyplot as plt
# line points
x_axis_line = [10,20,30,70,80,90]
#displaying axis limit
print(plt.axis())
y_axis_line= [20,40,10,23,22,12]
# plotting the line 1 points
plt.plot(x_axis_line, y_axis_line)
#setting title to given plot
plt.title("my first line")
#setting x axis lable
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
#setting new axis limit
#Set the y-limits of the current axes.
plt.ylim(0,100)
#Set the x-limits of the current axes.
plt.xlim(0,100)
# Display a figure.
plt.show()
```
