# plotting red wine data 

```python
#importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#loading data sets
red_wine = pd.read_csv('winequality-red.csv',sep='\;')
red_wine.head(10)
```

```python
red_wine.columns
```

```python
#taking columns seperately
density_red              = red_wine['"density"']
residual_sugar_red       = red_wine['"residual sugar"']
total_sulfur_dioxide_red = red_wine['"total sulfur dioxide"']
fixed_acidity_red        = red_wine['"fixed acidity"']
quqlity_red              = red_wine['"quality"']
```

plotting graphs

```python
#plotting complete data frame
red_wine.plot()
```

```python
#plotting bar graph
red_wine.plot(x='"quality"', y=['"residual sugar"', '"total sulfur dioxide"', '"fixed acidity"','"density"'], kind="bar")
```

```python
#plotting heatmap
sns.heatmap(red_wine.corr())
```

```python
#plotting subplots of all data 
bar_width = 0.35
opacity = 0.8
plt.bar(quqlity_red,density_red,bar_width,color='r',label='density')
plt.bar(quqlity_red+0.4,residual_sugar_red,bar_width,color='g',label='residual sugar')
plt.bar(quqlity_red+0.6,total_sulfur_dioxide_red,bar_width,color='b',label='residual sugar')
plt.bar(quqlity_red+0.8,fixed_acidity_red,bar_width,color='y',label='residual sugar')
plt.legend()
```

```python
#making subplots of line
axis_one = plt.subplot2grid((8,8),(0,0),rowspan=2,colspan=2)
axis_two = plt.subplot2grid((8,8),(0,4),rowspan=2,colspan=2)
axis_three = plt.subplot2grid((8,8),(4,0),rowspan=2,colspan=2)
axis_four = plt.subplot2grid((8,8),(4,4),rowspan=2,colspan=2)
axis_one.scatter(quqlity_red,density_red,s=0.1)
axis_two.scatter(quqlity_red,residual_sugar_red,s=0.1)
axis_three.scatter(quqlity_red,total_sulfur_dioxide_red,s=0.1)
axis_four.scatter(quqlity_red,fixed_acidity_red,s=0.1)
```

```python
#making subplots of bars
bar_one = plt.subplot2grid((8,8),(0,0),rowspan=2,colspan=2)
bar_two = plt.subplot2grid((8,8),(0,4),rowspan=2,colspan=2)
bar_three = plt.subplot2grid((8,8),(4,0),rowspan=2,colspan=2)
bar_four = plt.subplot2grid((8,8),(4,4),rowspan=2,colspan=2)
bar_one.bar(quqlity_red,density_red)
bar_two.bar(quqlity_red,residual_sugar_red)
bar_three.bar(quqlity_red,total_sulfur_dioxide_red)
bar_four.bar(quqlity_red,fixed_acidity_red)
```

```python
plt.hist(quqlity_red,facecolor='blue', alpha=0.5)
plt.hist(density_red,facecolor='green', alpha=0.5)
```

```python
sns.distplot(red_wine['"residual sugar"'], hist=False, rug=True)
sns.distplot(red_wine[['"total sulfur dioxide"']], hist=False, rug=True)
sns.distplot(red_wine[['"fixed acidity"']], hist=False, rug=True)
sns.distplot(red_wine[['"quality"']], hist=False, rug=True)

plt.show()
```
